```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from functools import reduce
```

```python
df = pd.read_csv('regression.csv')
df['score'].describe()
```

```python
def show_hists(df):
    fig, axs = plt.subplots(3,3, figsize=(18, 8))
    axs = list(reduce(lambda x, y: list(x) + list(y), axs))
    for ax, bins in zip(axs, ['auto', 10, 20, 30, 40, 50, 70, 80, 90]):
        ax.hist(df[['score']], bins=bins)
        ax.set_title(bins)
    plt.show()
show_hists(df)
```

```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df, test_size=0.25)
```

```python
show_hists(train_df)
```

```python
show_hists(test_df)
```

```python
train_df['name'] = 'Images/' + train_df['name']
test_df['name'] = 'Images/' + test_df['name']
```

```python
train_df.to_csv('train.csv')
test_df.to_csv('test.csv')
```
